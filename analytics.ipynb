```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("fitness_log.csv")
```

```python
data
```

```python
fig, ax = plt.subplots(figsize=(20,10))

ax.plot(data['avg_fitness'], label = 'avg fitness')
ax.set_ylabel('avg fitness')

ax2 = ax.twinx()
ax2.plot(data['best_fitness'], color='orange', label = 'best fitness')
ax2.set_ylabel('best fitness')

fig.legend()
plt.title("fitness over generations")
```
